```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import datetime as dt
import statsmodels.api as sm
```

```python
import pandas as pd


df = pd.read_csv("/content/drive/MyDrive/ML data/IndiaCovid19 - Sheet1.csv")


df['Date'] = pd.to_datetime(df['Date'])


training_start_date = pd.to_datetime('2020-05-01')
training_end_date = pd.to_datetime('2021-02-28')
evaluation_start_date = pd.to_datetime('2021-03-01')
evaluation_end_date = pd.to_datetime('2021-05-31')


training_data = df[(df['Date'] >= training_start_date) & (df['Date'] <= training_end_date)]
evaluation_data = df[(df['Date'] >= evaluation_start_date) & (df['Date'] <= evaluation_end_date)]


training_data.to_csv('training_data.csv', index=False)
evaluation_data.to_csv('evaluation_data.csv', index=False)
```

```python
model = LinearRegression()


X_train = training_data[['Daily Death Count']]
y_train = training_data[['StringencyIndex']]


model.fit(X_train, y_train)
```

```python
x_eval = evaluation_data[['Daily Death Count']]
y_eval = evaluation_data[['StringencyIndex']]
Predicted = model.predict(x_eval)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_eval, y_eval, label='ActualData')
plt.plot(x_eval, Predicted, color='red',linewidth=2, label='Regression Line')
plt.xlabel('Daily Death Count')
plt.ylabel('Predicted StringencyIndex')
plt.title('Linear Regression Prediction')
plt.legend()
plt.show()
```
